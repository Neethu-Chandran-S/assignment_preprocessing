# Check and handle outliers in at least 3 columns in the dataset


```python
import seaborn as sns

# Create seaborn box plot for 'Age' column
sns.boxplot(x='Age', data=titanic)
```

```python
##Z- score method of identifying outliers:
data = titanic['Age']

# Calculate the Z-scores
z_scores = (data - data.mean()) / data.std()
```

```python
z_scores.mean()
```

```python
# Define a threshold for identifying outliers (e.g., Z-score threshold of 2)
threshold = 2
# c                              NORMAL
#normal range: (outliers)  -2 ---------- 2 (outliers)
```

```python
# Identify the outliers
outliers = titanic[abs(z_scores) > threshold]

# Print the outliers
print("Outliers:")
outliers
```

```python
##IQR Method of handling Outliers
# Calculate the IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Define a threshold for identifying outliers (e.g., 1.5 times the IQR)
threshold = 1.5

# Identify the outliers
outliers = titanic[(data < Q1 - threshold * IQR) | (data > Q3 + threshold * IQR)]

# Print the outliers
print("Outliers:")
outliers
```

# Do min max scaling on the feature set (Take ‘Survived’ as target)


```python
import pandas as pd
##from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
titanic=pd.read_csv(r'C:\dsa\titanic_dataset.csv')
titanic.head()
```

```python
titanic_df=pd.DataFrame(data=titanic,columns=titanic['Survived'])

titanic_df["target"]=titanic['Survived']
titanic_df.head()
```

```python
X=titanic_df.drop('target',axis=1)
##create the MinMax scaler
scaler=MinMaxScaler()
scaler.fit(X)
X_normalized=scaler.transform(X)
print("\n Normalized Iris Dataset:")
result_norm=pd.DataFrame(data=X_normalized,columns=titanic_df["target"])
result_norm
```
